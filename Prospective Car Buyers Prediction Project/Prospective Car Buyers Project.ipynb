### ANALISIS PREDIKSI CALON PEMBELI MOBIL

#### LOGISTIK REGRESSION

```python
import pandas as pd

df1 = pd.read_csv('calonpembeli.csv')
```

Beli_Mobil
- 0 = tidak akan membeli mobil
- 1 = akan membeli mobil

```python
df1.describe()
```

```python
df1 = df1[df1['Usia'] <= 100]
```

```python
df1.isnull().sum()
```

```python
# Lihat berapa jumlah yang membeli dan tidak membeli

df1['Beli_Mobil'].value_counts()
```

```python
import sklearn.model_selection as ms

X=df1[['Usia', 'Status', 'Kelamin', 'Memiliki_Mobil', 'Penghasilan']]
y=df1.Beli_Mobil

X_train, y_train, X_test, y_test = ms.train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
import numpy as np
import sklearn.linear_model as lm

model = lm.LogisticRegression(solver='lbfgs')
model.fit(X_train, y_train)
```
